```python
# python program for basic arithmetic operations
#using 1 d array
import numpy as np
arr=np.array([1,3,5,7,9])
sum=arr[0]+arr[2]
print("sum of two numbers ",sum)
diff=arr[3]-arr[2]
print("difference between two numbers",diff)
product=arr[1]*arr[3]
div=arr[4]/arr[1]
print("product of two numbers",product)
print("division of two numbers",div)
```

```python
#slicing 1D array
import numpy as np
arr=np.array([2,4,6,8,12,14,16])
print(arr[3:6])
print(arr[:5])
print(arr[2:])
print(arr[-5:-1])
print(arr[:-1])


```

```python
#elements wise arithmetic operations
import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
print(arr1)
print(arr2)
print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
```

```python
# concatenation along the rows and columns
import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
print(arr1)
print(arr2)
concatenate_arr=np.concatenate((arr1,arr2),axis=0)# concatenate along thre rows
print(concatenate_arr)
concatenate_arr=np.concatenate((arr1,arr2),axis=1)# concatenate along the columns
print(concatenate_arr)
```
