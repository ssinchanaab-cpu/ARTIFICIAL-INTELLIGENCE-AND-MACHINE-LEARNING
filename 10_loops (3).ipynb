```python
for loop
Number of repetitions is known
Loop runs for a fixed range


while loop
Number of repetitions is unknown
Loop runs until a condition is satisfied

for loop is used when the number of iterations is known in advance,
whereas while loop is used when the number of iterations depends on a condition.
```

```python
# print numbers from 1 to 10
# We already know we must print numbers from 1 to 10.
# The number of repetitions is fixed (10 times)

for i in range(1, 11):
    print(i)
```

```python
#Accept the number from the user until the user enters 0
# as we do not know how many numbers the user will enter before entering 0
# go with while loop

num = 1
while num != 0:
    num = int(input("Enter number: "))
    print(num)
```

```python
# cant we write the above program using for loop?
# can but for loop requires guessing a limit and use break

for i in range(100):
    num = int(input("Enter number: "))
    if num == 0:
        break
```

```python
#Keep asking marks until the user enters a valid mark (0–100)
marks = -1

while marks < 0 or marks > 100:
    marks = int(input("Enter valid marks: "))
```

```python
#adding numbers until the sum becomes greater than 50.
#We do not know how many numbers are needed.

sum = 0
num = 1

while True:
    sum += num
    if sum > 50:
        break
    num += 1  # move to the next number

print("Sum exceeded 50")
print(sum)
```

```python
sum = 0
num = 1

while True:
    prev_sum = sum          # store the sum before adding
    sum += num              # add current number

    if sum > 50:
        print("Previous sum:", prev_sum)
        print("Current sum (exceeded 50):", sum)
        break

    num += 1                # move to next number
```

```python
sum = 0
num = 1

while sum <= 50:
    sum += num
    num += 1

print("Sum exceeded 50", sum)
```

```python
# cant we write the above program using for loop?
# can but need the guessing limit

sum = 0

for num in range(1, 100):   # 100 is an assumed value
    sum += num
    if sum > 50:
        break

print("Sum exceeded 50",sum)
```
