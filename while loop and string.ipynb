```python
crops=0
points=0     
print("1.seed is planted")
print("2.crop is watered")
print("3.crop is harvested")
print("4.exit")
while (True):
    choice=int(input("enter a choice"))
    if choice ==1:
        crops=crops+1
        print("seed is planted")
    elif choice==2:
         if crops>=1:
            print("crop is watered")
         else:
            print("no crop to water")
    elif choice==3:
        if crops>=1:
            points=points+(crops*10)
            crops=0
            print("crop is harvested")
        else:
            print("no crop to harvest")
    
    elif choice==4:
     print("total points",points)
     print("thanks for playing")
     break
    else:
        print("invalid choice")
            
```
